```python
# Split the dataset into features (x) and target (y)
x = df.drop('letter', axis=1)
x
```

```python
y = df['letter']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y.values.reshape(-1, 1))
y_onehot
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_onehot, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
model =Sequential()
model.add(layers.InputLayer(input_shape=(x_train.shape[1],)))  # Input layer
model.add(layers.Dense(64, activation='relu'))                   # Hidden layer 1
model.add(layers.Dense(32, activation='relu'))                   # Hidden layer 2
model.add(layers.Dense(y_onehot.shape[1], activation='softmax'))  # Output layer
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))
```

```python
val_loss, val_accuracy = model.evaluate(x_test, y_test)
print(f"Validation Accuracy: {val_accuracy:.4f}")
```

```python
predictions = model.predict(x_test)  # Predict on validation set
```

```python
predicted_classes = encoder.inverse_transform(predictions)  # Get class labels
print(predicted_classes)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(50),history.history['accuracy'],label='Train')
plt.plot(range(50),history.history['val_accuracy'],label='Test')
plt.legend(loc='best')
plt.xlabel('No of Epochs')
plt.ylabel('Acc and Val_Acc');
```

```python
plt.plot(range(50),history.history['loss'],label='Train')
plt.plot(range(50),history.history['val_loss'],label='Test')
plt.legend(loc='best')
plt.xlabel('No of Epochs')
plt.ylabel('loss and Val_loss');
```

### Discuss the performance differences between the model with default hyperparameters and the tuned model, emphasizing the effects of hyperparameter tuning.

When comparing a model with default hyperparameters and one with tuned hyperparameters, there are several key factors to consider in terms of performance. Hyperparameter tuning is crucial because the right set of parameters can significantly enhance a model's predictive ability and generalization to unseen data.

1. Model Accuracy and Error Metrics:
Default Model: Models with default hyperparameters are typically trained with a standard set of values chosen to work well for a wide range of problems. However, they might not be optimized for the specific characteristics of your dataset.

Accuracy: The accuracy of a model with default hyperparameters can be suboptimal, as the default settings may not match the data's inherent patterns.
Error Metrics: Common metrics (like MAE, RMSE, or MAPE) will likely show higher error values because the model may not be tuned to minimize the specific loss for the given data.
Tuned Model: Hyperparameter tuning involves adjusting key parameters (like learning rate, regularization strength, or the number of trees in a forest) to optimize model performance.

Accuracy: After hyperparameter tuning, the model is expected to perform better as it is customized to handle the nuances of the dataset. The tuning process helps to reduce overfitting/underfitting by selecting optimal values for the parameters.
Error Metrics: Tuned models often show improvements in error metrics because they have been adjusted to better fit the underlying patterns in the data. For instance, tuning may reduce both bias and variance, resulting in lower MAPE or RMSE values.
2. Generalization Ability:
Default Model: With the default settings, the model might struggle to generalize well on new or unseen data, as it hasn't been optimized to learn the best features of the dataset.
Tuned Model: Hyperparameter tuning aims to improve the model's ability to generalize by preventing overfitting and ensuring that the model doesn't learn noise in the training data. This leads to better performance on validation and test datasets.
3. Overfitting and Underfitting:
Default Model: The default model may be prone to overfitting if it has too many parameters relative to the dataset size or may suffer from underfitting if it doesn't have enough flexibility to capture the underlying patterns in the data.
Tuned Model: Tuning helps strike the right balance between bias and variance. Parameters like regularization, model complexity, or learning rate can be adjusted to prevent overfitting (by making the model simpler) or underfitting (by making the model more complex).
4. Training Time and Efficiency:
Default Model: Models with default hyperparameters are generally faster to train, as they don’t involve searching for optimal values. However, the results might not be as good.
Tuned Model: Hyperparameter tuning requires more computational resources and time, especially if techniques like grid search or random search are used. In some cases, tuning can significantly increase training time, but it is often justified by the gains in performance.
5. Hyperparameter Sensitivity:
Default Model: The performance of the default model might be relatively insensitive to the specific characteristics of the dataset because it uses standardized values.
Tuned Model: A model with tuned hyperparameters is more sensitive to the features and structure of the dataset, as tuning specifically adapts the model to fit the data.
Practical Example:
Let’s assume we are using an XGBoost model for classification:

Default Model: When using the default parameters for max_depth, learning_rate, and n_estimators, the model might show an accuracy of 80%, with a relatively high bias.
Tuned Model: After conducting hyperparameter tuning (e.g., using grid search), we might find the optimal parameters: max_depth=6, learning_rate=0.05, and n_estimators=1000. The accuracy could improve to 85%, with a better trade-off between bias and variance, leading to improved generalization.
