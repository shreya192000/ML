```python
Q1=df['LB'].quantile(0.25)
Q3=df['LB'].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
lower_extreme=Q1-1.5*IQR
```

```python
upper_extreme=Q3+1.5*IQR
```

```python
print(lower_extreme,upper_extreme)
```

```python
data_LB_outliers=data[(data['LB']<lower_extreme) | (df['LB']>upper_extreme)]
```

```python
data_LB_outliers
```

```python
data.boxplot('LB')
```

```python
data_no_outliers=data[(data['LB']>=lower_extreme) & (df['LB']<=upper_extreme)]
data_no_outliers
```
